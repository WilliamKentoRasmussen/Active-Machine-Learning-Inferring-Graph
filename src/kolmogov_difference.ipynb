```python
import numpy as np
from scipy import stats
```

```python
rng = np.random.default_rng()
rng
```

```python
x = stats.uniform.rvs(size=100, random_state=rng)
y = stats.uniform.rvs(size=100)
KstestResult = stats.kstest(x, y)
```

```python
p_value = KstestResult.statistic

if (p_value < 0.05): 
    print("The p-value of", p_value, " is below threshold of 0.05, so we can reject the null hypothesis: x can not be identical to y \n\n The null hypothesis is that the two distributions are identical\n")

else:
    print("The p-value of", p_value, "is above our threshold of 0.05, so we cannot reject the null hypothesis:  x could be identical to y \n\n The null hypothesis is that the two distributions are identical\n")
```
