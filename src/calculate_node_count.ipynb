```python
import math
from functools import lru_cache

@lru_cache(maxsize=None)
def a(N):
    """Compute a_N = sum_{k=1}^N (-1)^(k-1) * C(N,k) * 2^k * (N-k) * a_(N-k)"""
    if N == 1:
        return 1  # Base case - adjust as needed
    
    total = 0
    for k in range(1, N + 1):
        # Binomial coefficient C(N, k)
        binom = math.comb(N, k)
        
        # (-1)^(k-1)
        sign = (-1) ** (k - 1)
        
        term = sign * binom * (2 ** k) * (N - k) * a(N - k)
        total += term
    
    return total

# Example usage
for n in range(8):
    print(f"a_{n} = {a(n)}")
```
