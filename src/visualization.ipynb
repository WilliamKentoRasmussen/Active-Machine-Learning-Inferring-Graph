```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv('/Users/keremozemre/Library/CloudStorage/OneDrive-DanmarksTekniskeUniversitet/active_machine_learning/02463 Active machine learning and agency, Spring 2026 - 472026 - 1247 PM/Active-Machine-Learning-Inferring-Graph/DataTest/data_1451.csv')
data.info()
data = data.drop(columns=["Unnamed: 0"])
data["D"].mean()
```

```python
def plot_nodes(data):
    bins = 100
    plt.figure(figsize=(10, 4))
    plt.subplot(2, 2, 1)
    plt.hist(data["A"], bins=bins)
    plt.title("A")

    plt.subplot(2, 2, 2)
    plt.hist(data["B"], bins=bins)
    plt.title("B")

    plt.subplot(2, 2, 3)
    plt.hist(data["C"], bins=bins)  # clean discrete bins
    plt.title("C")

    plt.subplot(2, 2, 4)
    plt.hist(data["D"], bins=bins)  # clean discrete bins
    plt.title("D")

    plt.tight_layout()
    plt.show()
```
