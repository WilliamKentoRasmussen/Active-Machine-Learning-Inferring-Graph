```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('data_1451.csv')
data.info()
data = data.drop(columns=["Unnamed: 0"])
data.describe()
```

```python
def plot_nodes(data):
    bins = 100
    plt.figure(figsize=(10, 4))
    plt.subplot(2, 2, 1)
    plt.hist(data["A"], bins=bins)
    plt.title("A")

    plt.subplot(2, 2, 2)
    plt.hist(data["B"], bins=bins)
    plt.title("B")

    plt.subplot(2, 2, 3)
    plt.hist(data["C"], bins=bins)  # clean discrete bins
    plt.title("C")

    plt.subplot(2, 2, 4)
    plt.hist(data["D"], bins=bins)  # clean discrete bins
    plt.title("D")

    plt.tight_layout()
    plt.show()
```
