```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as snb
import os 
import seaborn as sns
```

```python
os.getcwd()
data_filepath = "../data/interventions/"
```

```python


data = pd.read_csv(data_filepath+'data.csv')
data.info()
data = data.drop(columns=["Unnamed: 0"])
data.describe();
```

```python
def plot_nodes(data):
    bins = 100
    plt.figure(figsize=(10, 4))
    plt.subplot(3, 2, 1)
    plt.hist(data["A"], bins=bins)
    plt.xlim((-2,2))
    plt.title("A")

    plt.subplot(3, 2, 2)
    plt.hist(data["B"], bins=bins)
    plt.title("B")
    plt.xlim((-2,2))

    plt.subplot(3, 2, 3)
    plt.hist(data["C"], bins=bins)  # clean discrete bins
    plt.title("C")
    plt.xlim((-2,2))

    plt.subplot(3, 2, 4)
    plt.hist(data["D"], bins=bins)  # clean discrete bins
    plt.title("D")
    plt.xlim((-2,2))

    plt.subplot(3, 2, 5)
    plt.hist(data["E"], bins=bins)  # clean discrete bins
    plt.title("E")
    plt.xlim((-2,2))

    plt.subplot(3, 2, 6)
    plt.hist(data["F"], bins=bins)  # clean discrete bins
    plt.title("F")
    plt.xlim((-2,2))

    plt.tight_layout()
    plt.show()
```

```python
data_R0 = pd.read_csv(data_filepath+'data_1481.csv')

data_R0 = data_R0.drop(columns=["Unnamed: 0"])
print(data_R0.describe())
plot_nodes(data=data_R0)
```

```python
data_S0 = pd.read_csv(data_filepath+'data_1482.csv')

data_S0 = data_S0.drop(columns=["Unnamed: 0"])
print(data_S0.describe())
plot_nodes(data=data_S0)
```
